```python
df['DATE_NORMALIZED'] = df['DATE'].apply(normalize_date)
# df.reset_index(drop=True)
```

```python
nan_dates = df[df['DATE_NORMALIZED'].isna()]
print(nan_dates['DATE_NORMALIZED'])
# df['DATE_NORMALIZED'].value_counts()
```

```python
df = df.dropna(subset=['DATE_NORMALIZED'])
df.reset_index(drop=True)
```

```python
# Visualize the distribution of the years (Title Date)
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='DATE_NORMALIZED', bins=50, kde=False, color='red', edgecolor='black')
plt.title('Distribution of Artworks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Artworks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming 'DATE_NORMALIZED' contains year as an integer
df['CENTURY'] = ((df['DATE_NORMALIZED'] // 100) + 1).astype(int)
```

```python
# Pivot table for heatmap
pivot_table = df.pivot_table(index='TYPE', columns='CENTURY', aggfunc='size', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap="YlGnBu")
plt.title('Heatmap of Painting Types by Year')
plt.xlim(6,20)
plt.xlabel('Century')
plt.ylabel('Type')
plt.show()
```

```python
# Pivot table for heatmap
pivot_table = df.pivot_table(index='SCHOOL', columns='CENTURY', aggfunc='size', fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap="YlGnBu")
plt.title('Heatmap of Painting Types by School')
plt.xlim(6,20)
plt.xlabel('Century')
plt.ylabel('School')
plt.show()
```

```python
# Plot distribution of types of paintings within each century
plt.figure(figsize=(14, 8))
sns.countplot(x='CENTURY', hue='TYPE', data=df, palette='viridis')
plt.title('Distribution of Artwork Types by Century')
plt.xlabel('Century')
plt.ylabel('Number of Artworks')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

```python
# Visualize the number of artworks by school (School)
plt.figure(figsize=(10, 6))
sns.countplot(y='SCHOOL', data=df, order=df['SCHOOL'].value_counts().index)
plt.title('Number of Artworks by School')
plt.xlabel('Number of Artworks')
plt.ylabel('School')
plt.show()
```

```python
pivot_table = df.pivot_table(index='SCHOOL', columns='CENTURY', aggfunc='size', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap='cividis', linewidths=.5)
plt.title('Distribution of Artworks by School and Century')
plt.ylabel('School')
plt.xlabel('Century')
plt.xticks(rotation=45)
plt.show()
```
